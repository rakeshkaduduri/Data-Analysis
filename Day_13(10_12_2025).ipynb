#✔ Cleaning Tasks
- Convert "TotalCharges" to numeric (dataset has spaces → become NaN).
- Fill missing numeric values using:
Median for tenure, MonthlyCharges, TotalCharges.
- Replace missing categorical values with "Unknown".


```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

path = '/content/drive/My Drive/Telco-Customer-Churn.csv'
dd = pd.read_csv(path)
```

```python
dd.head()
```

```python
dd.columns
```

```python
dd.info()
```

```python
# Convert "TotalCharges" to numeric (dataset has spaces → become NaN).
dd['TotalCharges'] = pd.to_numeric(dd['TotalCharges'], errors='coerce')
```

```python
dd.info()
```

```python
dd.isnull().sum()
```

```python
# Fill missing numeric values using: Median for tenure, MonthlyCharges, TotalCharges.
dd['tenure'].fillna(dd['tenure'].median())
dd['MonthlyCharges'].fillna(dd['MonthlyCharges'].median())
dd['TotalCharges'].fillna(dd['TotalCharges'].median(), inplace=True)
```

```python
dd.isnull().sum()
```

```python
dd[['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection']].fillna('Unknown')
```


#✔ Feature Engineering
Create the following new columns:
1. tenure_group
Based on tenure months:
   - 0–12   → "New"
   - 13–36  → "Regular"
   - 37–60  → "Loyal"
   - 60+    → "Champion"


```python
dd['tenure_group'] = pd.cut(dd['tenure'], bins=[0, 12, 36, 60, float('inf')], labels=['New', 'Regular', 'Loyal', 'Champion'])
```

```python
dd.head()
```

2. monthly_charge_segment
    - MonthlyCharges < 30  → "Low"
    - 30–70              → "Medium"
    - > 70                 → "High"


```python
dd['monthly_charge_segment'] = pd.cut(dd['MonthlyCharges'], bins=[0, 30, 70, float('inf')], labels=['Low', 'Medium', 'High'])
```

```python
dd.head()
```

3. has_internet_service
Convert InternetService column:
    - "DSL" / "Fiber optic" → 1
    - "No" → 0


```python
dd['has_internet_service'] = dd['InternetService'].apply(lambda x: 1 if x in ['DSL', 'Fiber optic'] else 0)
```

```python
dd.head()
```

4. is_multi_line_user
1 if MultipleLines == "Yes"
0 otherwise


```python
dd['is_multi_line_user'] = dd['MultipleLines'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
dd.head()
```

5. contract_type_code
- Map:
    - Month-to-month → 0
    - One year      → 1
    - Two year      → 2


```python
dd['contract_type_code'] = dd['Contract'].map({'Month-to-month': 0, 'One year': 1, 'Two year': 2})
```

```python
dd.head()
```

#✔ Drop unnecessary fields
Remove:
   - customerID, gender

```python
dd.drop(columns=['customerID', 'gender'], inplace=True)
```
